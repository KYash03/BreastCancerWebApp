```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/raw/data.csv')
```

```python
df.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[:, 1:].describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
X = df.iloc[:, 1:]
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.4,
                                                    random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
pipeline_svc = Pipeline([('scaler', StandardScaler()), ('svc_model', SVC())])
```

```python
pipeline_svc.fit(X_train, y_train)
```

```python
y_pred_svc = pipeline_svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred_svc)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
pipeline_knn = Pipeline([('scaler', StandardScaler()),
                         ('knn_model', KNeighborsClassifier())])
```

```python
pipeline_knn.fit(X_train, y_train)
```

```python
y_pred_knn = pipeline_knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_knn)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(pipeline_svc, X, y, cv=10).mean()
```

```python
cross_val_score(pipeline_knn, X, y, cv=10).mean()
```

#### I will optimize the SVC model.

```python
pipeline_svc['svc_model'].get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [{
    'svc_model__kernel': ['linear', 'poly'],
    'svc_model__C': np.arange(1, 100),
    'svc_model__degree': np.arange(1, 10)
}]
```

```python
svc_grid_search = GridSearchCV(
    Pipeline(steps=[('scaler', StandardScaler()), ('svc_model', SVC())]),
    parameters,
    n_jobs=-1)
```

```python
svc_grid_search.fit(X_train, y_train)
```

```python
svc_grid_search.best_params_
```

```python
updated_pipeline_svc = Pipeline(
    steps=[('scaler', StandardScaler()),
           ('svc_model',
            SVC(kernel=svc_grid_search.best_params_['svc_model__kernel'],
                C=svc_grid_search.best_params_['svc_model__C'],
                degree=svc_grid_search.best_params_['svc_model__degree']))])
```

```python
updated_pipeline_svc.fit(X_train, y_train)
```

```python
y_pred_svc_updated = updated_pipeline_svc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_svc_updated)
```

```python
cross_val_score(updated_pipeline_svc, X, y, cv=10).mean()
```
